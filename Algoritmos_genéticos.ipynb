```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import random
from deap import base, creator, tools, algorithms
import warnings
warnings.filterwarnings('ignore')

# Configurar estilo de gráficas
plt.style.use('default')
sns.set_palette("husl")

class MenuInteractivo:
    """Clase para manejar la interfaz de usuario"""

    def __init__(self):
        self.opciones = {
            '1': 'Maximización de Función Matemática',
            '2': 'Problema del Viajante (TSP)',
            '3': 'Comparación de Métodos de Selección',
            '4': 'Análisis de Parámetros',
            '5': 'Experimento con Funciones Personalizadas',
            '6': 'Implementación con DEAP (parámetros personalizables)',
            '7': 'Demostración de Mutación y Cruce (dinámica)',
            '8': 'Ejecutar Todo (Demo Completa)',
            '0': 'Salir'
        }

    def mostrar_menu(self):
        """Mostrar el menú principal"""
        print("\n" + "="*60)
        print("ALGORITMOS GENÉTICOS - MENÚ INTERACTIVO")
        print("="*60)
        for key, value in self.opciones.items():
            print(f"{key}. {value}")
        print("="*60)

    def obtener_seleccion(self):
        """Obtener la selección del usuario"""
        while True:
            try:
                seleccion = input("\nSelecciona una opción (0-8): ").strip()
                if seleccion in self.opciones:
                    return seleccion
                else:
                    print("Opción inválida. Por favor, selecciona un número del 0 al 8.")
            except KeyboardInterrupt:
                print("\n\n ¡Hasta luego!")
                return '0'
            except EOFError:
                return '0'

    def obtener_parametros_basicos(self):
        """Obtener parámetros básicos del usuario"""
        print("\nCONFIGURACIÓN DE PARÁMETROS")
        print("-" * 40)

        try:
            tam_poblacion = input("Tamaño de población (default 50): ").strip()
            tam_poblacion = int(tam_poblacion) if tam_poblacion else 50

            generaciones = input("Número de generaciones (default 100): ").strip()
            generaciones = int(generaciones) if generaciones else 100

            prob_cruce = input("Probabilidad de cruce 0-1 (default 0.7): ").strip()
            prob_cruce = float(prob_cruce) if prob_cruce else 0.7

            prob_mutacion = input("Probabilidad de mutación 0-1 (default 0.2): ").strip()
            prob_mutacion = float(prob_mutacion) if prob_mutacion else 0.2

            return {
                'tam_poblacion': max(10, tam_poblacion),
                'generaciones': max(10, generaciones),
                'prob_cruce': max(0.0, min(1.0, prob_cruce)),
                'prob_mutacion': max(0.0, min(1.0, prob_mutacion))
            }
        except (ValueError, KeyboardInterrupt):
            print("Valores inválidos o cancelado, usando parámetros por defecto.")
            return {
                'tam_poblacion': 50,
                'generaciones': 100,
                'prob_cruce': 0.7,
                'prob_mutacion': 0.2
            }

class AlgoritmoGeneticoOptimizado:
    """Clase optimizada para algoritmos genéticos"""

    def __init__(self, funcion_fitness, limite_inferior=-10, limite_superior=10,
                 tam_poblacion=50, probabilidad_cruce=0.7, probabilidad_mutacion=0.2,
                 verbose=True):
        self.funcion_fitness = funcion_fitness
        self.limite_inf = limite_inferior
        self.limite_sup = limite_superior
        self.tam_poblacion = tam_poblacion
        self.prob_cruce = probabilidad_cruce
        self.prob_mutacion = probabilidad_mutacion
        self.verbose = verbose
        self.historial_fitness = []
        self.mejor_global = None
        self.fitness_global = float('-inf')

    def crear_individuo(self):
        """Crear un individuo aleatorio"""
        return [random.uniform(self.limite_inf, self.limite_sup)]

    def crear_poblacion(self):
        """Crear población inicial"""
        return [self.crear_individuo() for _ in range(self.tam_poblacion)]

    def evaluar_fitness(self, individuo):
        """Evaluar fitness de un individuo"""
        return self.funcion_fitness(individuo[0])

    def seleccion_torneo(self, poblacion, fitness_scores, k=3):
        """Selección por torneo"""
        seleccionados = []
        for _ in range(len(poblacion)):
            torneo_indices = random.sample(range(len(poblacion)), k)
            ganador_idx = max(torneo_indices, key=lambda i: fitness_scores[i])
            seleccionados.append(poblacion[ganador_idx].copy())
        return seleccionados

    def cruce_aritmetico(self, padre1, padre2):
        """Cruce aritmético mejorado"""
        if random.random() < self.prob_cruce:
            alpha = random.uniform(0.2, 0.8)
            hijo1 = [alpha * padre1[0] + (1 - alpha) * padre2[0]]
            hijo2 = [(1 - alpha) * padre1[0] + alpha * padre2[0]]
            return hijo1, hijo2
        return padre1.copy(), padre2.copy()

    def mutacion_adaptiva(self, individuo, generacion, max_generaciones):
        """Mutación que se adapta según la generación"""
        if random.random() < self.prob_mutacion:
            intensidad = 1.0 - (generacion / max_generaciones) * 0.5
            individuo[0] += random.gauss(0, 0.5 * intensidad)
            individuo[0] = max(self.limite_inf, min(self.limite_sup, individuo[0]))
        return individuo

    def evolucionar(self, generaciones=100):
        """Proceso evolutivo principal mejorado"""
        poblacion = self.crear_poblacion()
        self.historial_fitness = []  # Reset historial

        if self.verbose:
            print(f"Iniciando evolución con {self.tam_poblacion} individuos por {generaciones} generaciones")

        for gen in range(generaciones):
            # Evaluar fitness
            fitness_scores = [self.evaluar_fitness(ind) for ind in poblacion]

            # Encontrar el mejor de esta generación
            mejor_idx = fitness_scores.index(max(fitness_scores))
            mejor_actual = poblacion[mejor_idx]
            fitness_actual = fitness_scores[mejor_idx]

            # Actualizar mejor global
            if fitness_actual > self.fitness_global:
                self.fitness_global = fitness_actual
                self.mejor_global = mejor_actual.copy()

            # Guardar estadísticas
            fitness_promedio = sum(fitness_scores) / len(fitness_scores)
            self.historial_fitness.append({
                'generacion': gen,
                'mejor_fitness': fitness_actual,
                'fitness_promedio': fitness_promedio,
                'mejor_individuo': mejor_actual[0],
                'diversidad': np.std([ind[0] for ind in poblacion])
            })

            # Mostrar progreso cada 20 generaciones
            if self.verbose and (gen % 20 == 0 or gen == generaciones - 1):
                print(f"Gen {gen:3d}: Mejor = {fitness_actual:.4f}, Promedio = {fitness_promedio:.4f}")

            # Selección
            poblacion = self.seleccion_torneo(poblacion, fitness_scores)

            # Cruce y mutación
            nueva_poblacion = []
            for i in range(0, len(poblacion), 2):
                padre1 = poblacion[i]
                padre2 = poblacion[i + 1] if i + 1 < len(poblacion) else poblacion[0]

                hijo1, hijo2 = self.cruce_aritmetico(padre1, padre2)
                hijo1 = self.mutacion_adaptiva(hijo1, gen, generaciones)
                hijo2 = self.mutacion_adaptiva(hijo2, gen, generaciones)

                nueva_poblacion.extend([hijo1, hijo2])

            poblacion = nueva_poblacion[:self.tam_poblacion]

        return self.historial_fitness

    def visualizar_resultados(self, titulo="Evolución del Algoritmo Genético"):
        """Crear visualización completa de resultados"""
        if not self.historial_fitness:
            print("No hay datos para visualizar. Ejecuta evolucionar() primero.")
            return

        df = pd.DataFrame(self.historial_fitness)

        fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))

        # Gráfica 1: Evolución del fitness
        ax1.plot(df['generacion'], df['mejor_fitness'], 'r-', linewidth=2,
                label='Mejor Fitness', marker='o', markersize=4)
        ax1.plot(df['generacion'], df['fitness_promedio'], 'b--', linewidth=2,
                label='Fitness Promedio', marker='s', markersize=4)
        ax1.set_xlabel('Generación')
        ax1.set_ylabel('Fitness')
        ax1.set_title('Evolución del Fitness')
        ax1.legend()
        ax1.grid(True, alpha=0.3)

        # Gráfica 2: Función y mejor solución
        x_vals = np.linspace(self.limite_inf, self.limite_sup, 1000)
        y_vals = [self.funcion_fitness(x) for x in x_vals]
        ax2.plot(x_vals, y_vals, 'g-', linewidth=2, label='Función objetivo')

        if self.mejor_global:
            mejor_x = self.mejor_global[0]
            mejor_y = self.funcion_fitness(mejor_x)
            ax2.plot(mejor_x, mejor_y, 'ro', markersize=12,
                    label=f'Mejor: x={mejor_x:.3f}')

        ax2.set_xlabel('x')
        ax2.set_ylabel('f(x)')
        ax2.set_title('Función Objetivo y Mejor Solución')
        ax2.legend()
        ax2.grid(True, alpha=0.3)

        # Gráfica 3: Diversidad de la población
        ax3.plot(df['generacion'], df['diversidad'], 'purple', linewidth=2, marker='d')
        ax3.set_xlabel('Generación')
        ax3.set_ylabel('Diversidad (Desviación Estándar)')
        ax3.set_title('Diversidad de la Población')
        ax3.grid(True, alpha=0.3)

        # Gráfica 4: Convergencia del mejor individuo
        ax4.scatter(df['generacion'], df['mejor_individuo'],
                   c=df['generacion'], cmap='viridis', alpha=0.7)
        ax4.set_xlabel('Generación')
        ax4.set_ylabel('Valor x del Mejor Individuo')
        ax4.set_title('Convergencia del Mejor Individuo')
        ax4.grid(True, alpha=0.3)

        plt.colorbar(ax4.collections[0], ax=ax4, label='Generación')
        plt.suptitle(titulo, fontsize=16, fontweight='bold')
        plt.tight_layout()
        plt.show()
        plt.close()  # Cerrar la figura para liberar memoria

class TSPOptimizado:
    """Clase optimizada para el Problema del Viajante"""

    def __init__(self, ciudades, tam_poblacion=100, verbose=True):
        self.ciudades = ciudades
        self.n_ciudades = len(ciudades)
        self.tam_poblacion = tam_poblacion
        self.verbose = verbose
        self.distancias = self.calcular_matriz_distancias()
        self.nombres_ciudades = [chr(65 + i) for i in range(self.n_ciudades)]

    def calcular_matriz_distancias(self):
        """Calcular matriz de distancias euclidiana"""
        n = self.n_ciudades
        matriz = np.zeros((n, n))
        for i in range(n):
            for j in range(n):
                if i != j:
                    dx = self.ciudades[i][0] - self.ciudades[j][0]
                    dy = self.ciudades[i][1] - self.ciudades[j][1]
                    matriz[i][j] = np.sqrt(dx**2 + dy**2)
        return matriz

    def crear_individuo(self):
        """Crear una ruta aleatoria"""
        ruta = list(range(self.n_ciudades))
        random.shuffle(ruta)
        return ruta

    def calcular_distancia_total(self, ruta):
        """Calcular distancia total de una ruta"""
        distancia = 0
        for i in range(len(ruta)):
            ciudad_actual = ruta[i]
            ciudad_siguiente = ruta[(i + 1) % len(ruta)]
            distancia += self.distancias[ciudad_actual][ciudad_siguiente]
        return distancia

    def calcular_fitness(self, ruta):
        """Calcular fitness (inverso de la distancia)"""
        distancia = self.calcular_distancia_total(ruta)
        return 1 / (1 + distancia)

    def cruce_orden_mejorado(self, padre1, padre2):
        """Cruce por orden mejorado (OX)"""
        n = len(padre1)
        punto1, punto2 = sorted(random.sample(range(n), 2))

        hijo = [-1] * n
        hijo[punto1:punto2] = padre1[punto1:punto2]

        ptr = punto2
        for ciudad in padre2[punto2:] + padre2[:punto2]:
            if ciudad not in hijo:
                hijo[ptr % n] = ciudad
                ptr += 1

        return hijo

    def mutacion_2opt(self, ruta):
        """Mutación 2-opt (más efectiva para TSP)"""
        if random.random() < 0.2:
            i, j = sorted(random.sample(range(len(ruta)), 2))
            ruta[i:j+1] = reversed(ruta[i:j+1])
        return ruta

    def resolver(self, generaciones=200):
        """Resolver TSP usando algoritmo genético optimizado"""
        poblacion = [self.crear_individuo() for _ in range(self.tam_poblacion)]
        historial = []
        mejor_global = None
        mejor_distancia_global = float('inf')

        if self.verbose:
            print(f"Resolviendo TSP con {self.n_ciudades} ciudades...")

        for gen in range(generaciones):
            # Evaluar fitness
            fitness_scores = [self.calcular_fitness(ruta) for ruta in poblacion]
            distancias = [self.calcular_distancia_total(ruta) for ruta in poblacion]

            # Encontrar el mejor
            mejor_idx = fitness_scores.index(max(fitness_scores))
            mejor_ruta = poblacion[mejor_idx]
            mejor_distancia = distancias[mejor_idx]

            if mejor_distancia < mejor_distancia_global:
                mejor_distancia_global = mejor_distancia
                mejor_global = mejor_ruta.copy()

            historial.append({
                'generacion': gen,
                'mejor_distancia': mejor_distancia,
                'distancia_promedio': np.mean(distancias),
                'mejor_ruta': mejor_ruta.copy()
            })

            if self.verbose and (gen % 50 == 0 or gen == generaciones - 1):
                print(f"Gen {gen:3d}: Mejor distancia = {mejor_distancia:.2f}")

            # Selección por torneo
            nueva_poblacion = []
            for _ in range(self.tam_poblacion):
                torneo = random.sample(list(zip(poblacion, fitness_scores)), 5)
                ganador = max(torneo, key=lambda x: x[1])[0]
                nueva_poblacion.append(ganador.copy())

            # Cruce y mutación
            poblacion_hijos = []
            for i in range(0, len(nueva_poblacion), 2):
                padre1 = nueva_poblacion[i]
                padre2 = nueva_poblacion[i+1] if i+1 < len(nueva_poblacion) else nueva_poblacion[0]

                hijo1 = self.cruce_orden_mejorado(padre1, padre2)
                hijo2 = self.cruce_orden_mejorado(padre2, padre1)

                hijo1 = self.mutacion_2opt(hijo1)
                hijo2 = self.mutacion_2opt(hijo2)

                poblacion_hijos.extend([hijo1, hijo2])

            poblacion = poblacion_hijos[:self.tam_poblacion]

        return historial, mejor_global

    def visualizar_resultado(self, historial, mejor_ruta):
        """Visualizar resultado del TSP"""
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

        # Gráfica 1: Evolución de la distancia
        distancias = [h['mejor_distancia'] for h in historial]
        ax1.plot(distancias, 'b-', linewidth=2, marker='o', markersize=4)
        ax1.set_xlabel('Generación')
        ax1.set_ylabel('Distancia Total')
        ax1.set_title('Evolución de la Mejor Distancia')
        ax1.grid(True, alpha=0.3)

        # Gráfica 2: Mejor ruta
        ax2.scatter([c[0] for c in self.ciudades], [c[1] for c in self.ciudades],
                   c='red', s=200, alpha=0.7, zorder=5)

        # Dibujar la ruta
        ruta_x = [self.ciudades[i][0] for i in mejor_ruta] + [self.ciudades[mejor_ruta[0]][0]]
        ruta_y = [self.ciudades[i][1] for i in mejor_ruta] + [self.ciudades[mejor_ruta[0]][1]]
        ax2.plot(ruta_x, ruta_y, 'b-', linewidth=2, alpha=0.7)
        ax2.plot(ruta_x, ruta_y, 'bo', markersize=8, alpha=0.5)

        # Etiquetas de ciudades
        for i, (x, y) in enumerate(self.ciudades):
            ax2.annotate(self.nombres_ciudades[i], (x, y), xytext=(5, 5),
                        textcoords='offset points', fontsize=12, fontweight='bold')

        ax2.set_xlabel('X')
        ax2.set_ylabel('Y')
        ax2.set_title(f'Mejor Ruta - Distancia: {distancias[-1]:.2f}')
        ax2.grid(True, alpha=0.3)
        ax2.set_aspect('equal', adjustable='box')

        plt.tight_layout()
        plt.show()
        plt.close()

        print(f"Mejor ruta: {[self.nombres_ciudades[i] for i in mejor_ruta]}")
        print(f"Distancia total: {distancias[-1]:.2f}")
        mejora = ((distancias[0] - distancias[-1])/distancias[0]*100)
        print(f"Mejora: {mejora:.1f}%")

class GestorFunciones:
    """Gestor de funciones objetivo disponibles"""

    @staticmethod
    def obtener_funciones():
        return {
            1: {
                'nombre': 'f(x) = -(x-3)² + 10',
                'funcion': lambda x: -(x-3)**2 + 10,
                'limites': (-10, 10),
                'optimo_teorico': (3, 10)
            },
            2: {
                'nombre': 'f(x) = sin(x) + cos(2x)',
                'funcion': lambda x: np.sin(x) + np.cos(2*x),
                'limites': (-2*np.pi, 2*np.pi),
                'optimo_teorico': None
            },
            3: {
                'nombre': 'f(x) = e^(-x²/10) * cos(x)',
                'funcion': lambda x: np.exp(-x**2/10) * np.cos(x),
                'limites': (-5, 5),
                'optimo_teorico': None
            },
            4: {
                'nombre': 'f(x) = x * sin(10π*x) + 1',
                'funcion': lambda x: x * np.sin(10 * np.pi * x) + 1,
                'limites': (-1, 2),
                'optimo_teorico': None
            }
        }

def pausa_para_continuar(mensaje="Presiona Enter para continuar..."):
    """Función helper para pausar y esperar input del usuario"""
    try:
        input(f"\n {mensaje}")
    except (KeyboardInterrupt, EOFError):
        pass

def ejecutar_opcion_1(parametros):
    """Ejecutar maximización de función matemática"""
    try:
        print("\n MAXIMIZACIÓN DE FUNCIÓN MATEMÁTICA")
        print("="*50)

        funciones = GestorFunciones.obtener_funciones()

        print("Funciones disponibles:")
        for key, info in funciones.items():
            print(f"{key}. {info['nombre']}")

        try:
            seleccion_str = input("\n Selecciona una función (1-4): ").strip()
            seleccion = int(seleccion_str) if seleccion_str else 1
            if seleccion not in funciones:
                seleccion = 1
                print(f" Opción inválida, usando función {seleccion}")
        except (ValueError, KeyboardInterrupt):
            seleccion = 1
            print(f" Usando función por defecto {seleccion}")

        func_info = funciones[seleccion]

        ag = AlgoritmoGeneticoOptimizado(
            func_info['funcion'],
            func_info['limites'][0],
            func_info['limites'][1],
            parametros['tam_poblacion'],
            parametros['prob_cruce'],
            parametros['prob_mutacion']
        )

        print(f"\n Optimizando: {func_info['nombre']}")
        historial = ag.evolucionar(parametros['generaciones'])
        ag.visualizar_resultados(f"Optimización: {func_info['nombre']}")

        print(f"\n RESULTADOS:")
        print(f"   Mejor solución: x = {ag.mejor_global[0]:.4f}")
        print(f"   Valor óptimo: f({ag.mejor_global[0]:.4f}) = {ag.fitness_global:.4f}")

        if func_info['optimo_teorico']:
            opt_x, opt_y = func_info['optimo_teorico']
            print(f"   Óptimo teórico: f({opt_x}) = {opt_y}")
            error = abs(ag.mejor_global[0] - opt_x)
            print(f"   Error: {error:.4f}")

        print(" Maximización completada exitosamente")

    except Exception as e:
        print(f" Error en maximización de función: {str(e)}")

def ejecutar_opcion_2(parametros):
    """Ejecutar problema del viajante"""
    try:
        print("\n PROBLEMA DEL VIAJANTE (TSP)")
        print("="*40)

        opciones_ciudades = {
            1: {"n": 6, "nombre": "6 ciudades (fácil)"},
            2: {"n": 8, "nombre": "8 ciudades (medio)"},
            3: {"n": 10, "nombre": "10 ciudades (difícil)"},
            4: {"n": 12, "nombre": "12 ciudades (muy difícil)"}
        }

        print("Opciones disponibles:")
        for key, info in opciones_ciudades.items():
            print(f"{key}. {info['nombre']}")

        try:
            seleccion_str = input("\n Selecciona dificultad (1-4): ").strip()
            seleccion = int(seleccion_str) if seleccion_str else 2
            if seleccion not in opciones_ciudades:
                seleccion = 2
                print(f" Opción inválida, usando opción {seleccion}")
        except (ValueError, KeyboardInterrupt):
            seleccion = 2
            print(f" Usando opción por defecto {seleccion}")

        n_ciudades = opciones_ciudades[seleccion]["n"]

        # Generar ciudades aleatorias
        np.random.seed(42)
        ciudades = [(np.random.rand()*100, np.random.rand()*100) for _ in range(n_ciudades)]

        print(f" Resolviendo TSP con {n_ciudades} ciudades...")
        tsp = TSPOptimizado(ciudades, parametros['tam_poblacion'])
        historial, mejor_ruta = tsp.resolver(parametros['generaciones'])
        tsp.visualizar_resultado(historial, mejor_ruta)

        print(" TSP completado exitosamente")

    except Exception as e:
        print(f" Error en TSP: {str(e)}")

def ejecutar_comparacion_selecciones():
    """Comparar métodos de selección"""
    try:
        print("\n COMPARACIÓN DE MÉTODOS DE SELECCIÓN")
        print("="*50)

        def seleccion_ruleta(poblacion, fitness_scores):
            """Selección por ruleta"""
            min_fitness = min(fitness_scores)
            fitness_ajustado = [f - min_fitness + 1 for f in fitness_scores]
            total_fitness = sum(fitness_ajustado)

            seleccionados = []
            for _ in range(len(poblacion)):
                r = random.uniform(0, total_fitness)
                acumulado = 0
                for i, fitness in enumerate(fitness_ajustado):
                    acumulado += fitness
                    if acumulado >= r:
                        seleccionados.append(poblacion[i].copy())
                        break
            return seleccionados

        # Función de prueba
        funcion_test = lambda x: -(x-3)**2 + 10

        # Crear población de prueba
        poblacion_test = [[random.uniform(-5, 5)] for _ in range(20)]
        fitness_test = [funcion_test(ind[0]) for ind in poblacion_test]

        # Crear AG para usar selección por torneo
        ag_test = AlgoritmoGeneticoOptimizado(funcion_test, verbose=False)

        # Comparar métodos
        metodos = {
            'Ruleta': lambda: seleccion_ruleta(poblacion_test, fitness_test),
            'Torneo': lambda: ag_test.seleccion_torneo(poblacion_test, fitness_test)
        }

        print(" Comparando métodos de selección...")
        fig, axes = plt.subplots(1, 2, figsize=(15, 5))

        for idx, (nombre, metodo) in enumerate(metodos.items()):
            conteos = {}
            for _ in range(1000):
                seleccionados = metodo()
                for ind in seleccionados:
                    key = round(ind[0], 1)
                    conteos[key] = conteos.get(key, 0) + 1

            x_vals = sorted(conteos.keys())
            y_vals = [conteos[x] for x in x_vals]

            axes[idx].bar(x_vals, y_vals, alpha=0.7, color=['red', 'blue'][idx])
            axes[idx].set_title(f'Distribución - Selección {nombre}')
            axes[idx].set_xlabel('Valor x')
            axes[idx].set_ylabel('Frecuencia de selección')
            axes[idx].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()
        plt.close()

        print(" ANÁLISIS:")
        print("   • Selección por ruleta: Favorece más los mejores")
        print("   • Selección por torneo: Mantiene más diversidad")
        print(" Comparación completada")

    except Exception as e:
        print(f" Error en comparación de selecciones: {str(e)}")

def ejecutar_analisis_parametros():
    """Analizar efecto de diferentes parámetros"""
    try:
        print("\n ANÁLISIS DE PARÁMETROS")
        print("="*35)

        funcion_test = lambda x: -(x-3)**2 + 10

        # Analizar tamaño de población
        print("Analizando tamaño de población...")
        tamaños = [20, 50, 100, 150]
        resultados_tam = []

        for tam in tamaños:
            ag = AlgoritmoGeneticoOptimizado(funcion_test, tam_poblacion=tam, verbose=False)
            ag.evolucionar(30)
            resultados_tam.append(ag.fitness_global)
            print(f"  Población {tam:3d}: {ag.fitness_global:.4f}")

        # Analizar probabilidad de mutación
        print("\n Analizando probabilidad de mutación...")
        prob_mutaciones = [0.01, 0.1, 0.2, 0.4]
        resultados_mut = []

        for prob in prob_mutaciones:
            ag = AlgoritmoGeneticoOptimizado(funcion_test, probabilidad_mutacion=prob, verbose=False)
            ag.evolucionar(30)
            resultados_mut.append(ag.fitness_global)
            print(f"  P(mutación) {prob:4.2f}: {ag.fitness_global:.4f}")

        # Visualizar
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

        ax1.plot(tamaños, resultados_tam, 'bo-', linewidth=2, markersize=8)
        ax1.set_xlabel('Tamaño de Población')
        ax1.set_ylabel('Mejor Fitness')
        ax1.set_title('Efecto del Tamaño de Población')
        ax1.grid(True, alpha=0.3)

        ax2.plot(prob_mutaciones, resultados_mut, 'ro-', linewidth=2, markersize=8)
        ax2.set_xlabel('Probabilidad de Mutación')
        ax2.set_ylabel('Mejor Fitness')
        ax2.set_title('Efecto de la Probabilidad de Mutación')
        ax2.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.show()
        plt.close()

        print("CONCLUSIONES:")
        print(" • Población más grande generalmente mejora resultados")
        print(" • Mutación moderada (0.1-0.2) suele ser óptima")
        print("Análisis de parámetros completado")

    except Exception as e:
        print(f"Error en análisis de parámetros: {str(e)}")

def ejecutar_experimento_personalizado():
    """Experimento con funciones personalizadas"""
    try:
        print("\n EXPERIMENTO CON FUNCIONES PERSONALIZADAS")
        print("="*50)

        funciones = GestorFunciones.obtener_funciones()

        print("Selecciona múltiples funciones para comparar:")
        for key, info in funciones.items():
            print(f"{key}. {info['nombre']}")

        try:
            selecciones = input("\nIngresa números separados por comas (ej: 1,2,3): ").strip()
            if selecciones:
                indices = [int(x.strip()) for x in selecciones.split(',') if x.strip().isdigit() and int(x.strip()) in funciones]
            else:
                indices = []
            if not indices:
                indices = [1, 2]
                print("Usando funciones por defecto: 1, 2")
        except (ValueError, KeyboardInterrupt):
            indices = [1, 2]
            print("Usando funciones por defecto: 1, 2")

        print(f"\nComparando {len(indices)} funciones...")

        resultados = {}
        colores = ['red', 'blue', 'green', 'orange', 'purple']

        # Crear subplots para funciones
        n_funciones = len(indices)
        cols = 2
        rows = (n_funciones + 1) // 2

        fig = plt.figure(figsize=(15, 5 * rows))

        for i, idx in enumerate(indices):
            func_info = funciones[idx]

            ag = AlgoritmoGeneticoOptimizado(
                func_info['funcion'],
                func_info['limites'][0],
                func_info['limites'][1],
                tam_poblacion=50,
                verbose=False
            )

            historial = ag.evolucionar(100)
            resultados[idx] = {
                'historial': historial,
                'mejor_fitness': ag.fitness_global,
                'mejor_x': ag.mejor_global[0],
                'nombre': func_info['nombre']
            }

            # Subplot para cada función
            plt.subplot(rows, cols, i+1)

            # Graficar función
            x_vals = np.linspace(func_info['limites'][0], func_info['limites'][1], 1000)
            y_vals = [func_info['funcion'](x) for x in x_vals]
            plt.plot(x_vals, y_vals, 'g-', linewidth=2, alpha=0.7)

            # Marcar mejor solución
            plt.plot(ag.mejor_global[0], ag.fitness_global, 'ro', markersize=10)

            plt.title(func_info['nombre'])
            plt.xlabel('x')
            plt.ylabel('f(x)')
            plt.grid(True, alpha=0.3)

            print(f"  {func_info['nombre']}: x = {ag.mejor_global[0]:.4f}, f(x) = {ag.fitness_global:.4f}")

        plt.tight_layout()
        plt.show()
        plt.close()

        # Gráfica comparativa de evolución
        plt.figure(figsize=(12, 6))
        for i, idx in enumerate(indices):
            df = pd.DataFrame(resultados[idx]['historial'])
            plt.plot(df['generacion'], df['mejor_fitness'],
                    color=colores[i % len(colores)], linewidth=2,
                    label=f"Función {idx}", marker='o', markersize=4)

        plt.xlabel('Generación')
        plt.ylabel('Mejor Fitness')
        plt.title('Comparación de Evolución entre Funciones')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()
        plt.close()

        print("Experimento personalizado completado")

    except Exception as e:
        print(f"Error en experimento personalizado: {str(e)}")

def ejecutar_deap_ejemplo(auto=False):
    """Ejemplo usando DEAP con parámetros personalizables por el usuario"""
    try:
        print("\nIMPLEMENTACIÓN CON DEAP")
        print("="*35)

        # ----- NUEVO: pedir parámetros al usuario (si no es modo auto) -----
        if not auto:
            try:
                tam_poblacion = int(input("Tamaño de población (default 50): ") or 50)
                ngen = int(input("Número de generaciones (default 50): ") or 50)
                cxpb = float(input("Probabilidad de cruce (0-1, default 0.7): ") or 0.7)
                mutpb = float(input("Probabilidad de mutación (0-1, default 0.2): ") or 0.2)
                tournsize = int(input("Tamaño de torneo (default 3): ") or 3)
                lim_inf = float(input("Límite inferior del rango (default -10): ") or -10)
                lim_sup = float(input("Límite superior del rango (default 10): ") or 10)
            except (ValueError, KeyboardInterrupt):
                print("Parámetros inválidos. Usando valores por defecto.")
                tam_poblacion, ngen, cxpb, mutpb, tournsize, lim_inf, lim_sup = 50, 50, 0.7, 0.2, 3, -10, 10
        else:
            tam_poblacion, ngen, cxpb, mutpb, tournsize, lim_inf, lim_sup = 50, 50, 0.7, 0.2, 3, -10, 10

        # Limpiar creators existentes si existen
        try:
            if hasattr(creator, "FitnessMax"):
                del creator.FitnessMax
            if hasattr(creator, "Individual"):
                del creator.Individual
        except:
            pass

        # Configurar DEAP
        creator.create("FitnessMax", base.Fitness, weights=(1.0,))
        creator.create("Individual", list, fitness=creator.FitnessMax)

        toolbox = base.Toolbox()

        # Función objetivo
        def evaluar_deap(individual):
            x = individual[0]
            return (-(x-3)**2 + 10,)

        # Configurar herramientas
        toolbox.register("attr_float", random.uniform, lim_inf, lim_sup)
        toolbox.register("individual", tools.initRepeat, creator.Individual,
                         toolbox.attr_float, n=1)
        toolbox.register("population", tools.initRepeat, list, toolbox.individual)
        toolbox.register("evaluate", evaluar_deap)
        toolbox.register("mate", tools.cxBlend, alpha=0.5)
        toolbox.register("mutate", tools.mutGaussian, mu=0, sigma=1, indpb=0.2)
        toolbox.register("select", tools.selTournament, tournsize=tournsize)

        # Parámetros
        poblacion = toolbox.population(n=tam_poblacion)

        print("\nEjecutando algoritmo con DEAP...")
        print(f"Población={tam_poblacion}, Generaciones={ngen}, cxpb={cxpb}, mutpb={mutpb}, tournsize={tournsize}")

        # Estadísticas
        stats = tools.Statistics(lambda ind: ind.fitness.values)
        stats.register("avg", np.mean)
        stats.register("std", np.std)
        stats.register("min", np.min)
        stats.register("max", np.max)

        # Ejecutar evolución
        poblacion_final, logbook = algorithms.eaSimple(
            poblacion, toolbox, cxpb, mutpb, ngen,
            stats=stats, verbose=False
        )

        # Mejor individuo
        mejor_ind = tools.selBest(poblacion_final, 1)[0]

        print(f"\nRESULTADOS DEAP:")
        print(f"Mejor individuo: x = {mejor_ind[0]:.4f}")
        print(f"Mejor fitness: {mejor_ind.fitness.values[0]:.4f}")

        # Visualizar
        gen = logbook.select("gen")
        avg = logbook.select("avg")
        max_vals = logbook.select("max")

        plt.figure(figsize=(10, 6))
        plt.plot(gen, avg, 'b-', label='Promedio', linewidth=2, marker='s')
        plt.plot(gen, max_vals, 'r-', label='Máximo', linewidth=2, marker='o')
        plt.xlabel('Generación')
        plt.ylabel('Fitness')
        plt.title('Evolución con DEAP - Estadísticas Detalladas')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.show()
        plt.close()

        print("Ejemplo DEAP completado")

    except Exception as e:
        print(f"Error en ejemplo DEAP: {str(e)}")

def demostrar_mutacion_cruce(auto=False):
    """Demostración visual de operadores genéticos (dinámica)"""
    try:
        print("\nDEMOSTRACIÓN: MUTACIÓN Y CRUCE (DINÁMICA)")
        print("="*45)

        # ----- NUEVO: permitir al usuario definir longitud, punto de cruce y posición de mutación -----
        if not auto:
            try:
                longitud = int(input("Longitud del cromosoma binario (default 8): ") or 8)
                especificar = input("¿Quieres especificar punto de cruce y posición de mutación? (s/N): ").strip().lower() in ['s', 'si', 'y', 'yes']
                if especificar:
                    punto_cruce = int(input(f"Punto de cruce (1-{longitud-1}, default aleatorio): ") or random.randint(1, longitud-1))
                    pos_mut = int(input(f"Posición de mutación (0-{longitud-1}, default aleatorio): ") or random.randint(0, longitud-1))
                else:
                    punto_cruce = random.randint(1, longitud-1)
                    pos_mut = random.randint(0, longitud-1)
            except (ValueError, KeyboardInterrupt):
                print("Parámetros inválidos. Usando valores por defecto.")
                longitud = 8
                punto_cruce = random.randint(1, longitud-1)
                pos_mut = random.randint(0, longitud-1)
        else:
            longitud = 8
            punto_cruce = random.randint(1, longitud-1)
            pos_mut = random.randint(0, longitud-1)

        # Generar padres aleatorios
        padre1 = [random.randint(0, 1) for _ in range(longitud)]
        padre2 = [random.randint(0, 1) for _ in range(longitud)]

        # CRUCE EN UN PUNTO
        hijo1 = padre1[:punto_cruce] + padre2[punto_cruce:]
        hijo2 = padre2[:punto_cruce] + padre1[punto_cruce:]

        print("CRUCE EN UN PUNTO:")
        print(f"Padre 1:  {''.join(map(str, padre1))} -> {''.join(map(str, padre1[:punto_cruce]))}|{''.join(map(str, padre1[punto_cruce:]))}")
        print(f"Padre 2:  {''.join(map(str, padre2))} -> {''.join(map(str, padre2[:punto_cruce]))}|{''.join(map(str, padre2[punto_cruce:]))}")
        print(f"          {' ' * (punto_cruce + 9)}↑ Punto de cruce [{punto_cruce}]")
        print(f"Hijo 1:   {''.join(map(str, hijo1))}")
        print(f"Hijo 2:   {''.join(map(str, hijo2))}")

        # MUTACIÓN
        print(f"\nMUTACIÓN:")
        original = hijo1.copy()
        mutado = original.copy()
        mutado[pos_mut] = 1 - mutado[pos_mut]

        print(f"Original: {''.join(map(str, original))}")
        print(f"Mutado:   {''.join(map(str, mutado[:pos_mut]))}[{mutado[pos_mut]}]{''.join(map(str, mutado[pos_mut+1:]))}")
        print(f"          {' ' * pos_mut}↑ Mutación en posición {pos_mut}")

        # Demostración visual con gráfica
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

        # Visualizar cruce
        generations = ['Padre 1', 'Padre 2', 'Hijo 1', 'Hijo 2']
        chromosomes = [padre1, padre2, hijo1, hijo2]
        colors = ['lightblue', 'lightgreen', 'lightyellow', 'lightpink']

        for i, (gen, chrom, color) in enumerate(zip(generations, chromosomes, colors)):
            for j, bit in enumerate(chrom):
                ax1.add_patch(plt.Rectangle((j, i), 1, 0.8,
                                          facecolor=color,
                                          edgecolor='black',
                                          alpha=0.7))
                ax1.text(j+0.5, i+0.4, str(bit),
                        ha='center', va='center', fontsize=12, fontweight='bold')

        ax1.set_xlim(0, longitud)
        ax1.set_ylim(0, 4)
        ax1.set_yticks(range(4))
        ax1.set_yticklabels(generations)
        ax1.set_xticks(range(longitud))
        ax1.set_title('Demostración de Cruce')
        ax1.axvline(x=punto_cruce, color='red', linestyle='--', linewidth=2, alpha=0.7)

        # Visualizar mutación
        for i, (label, chrom) in enumerate([('Original', original), ('Mutado', mutado)]):
            color = 'lightblue' if i == 0 else 'lightcoral'
            for j, bit in enumerate(chrom):
                special_color = 'red' if i == 1 and j == pos_mut else color
                ax2.add_patch(plt.Rectangle((j, i), 1, 0.8,
                                          facecolor=special_color,
                                          edgecolor='black',
                                          alpha=0.7))
                ax2.text(j+0.5, i+0.4, str(bit),
                        ha='center', va='center', fontsize=12, fontweight='bold')

        ax2.set_xlim(0, longitud)
        ax2.set_ylim(0, 2)
        ax2.set_yticks(range(2))
        ax2.set_yticklabels(['Original', 'Mutado'])
        ax2.set_xticks(range(longitud))
        ax2.set_title('Demostración de Mutación')

        plt.tight_layout()
        plt.show()
        plt.close()

        print("CONCEPTOS CLAVE:")
        print(" • Cruce: Combina material genético de dos padres")
        print(" • Mutación: Introduce variación aleatoria")
        print(" • Ambos son esenciales para la exploración")
        print("Demostración completada")

    except Exception as e:
        print(f"Error en demostración: {str(e)}")

def ejecutar_demo_completa():
    """Ejecutar demostración completa"""
    try:
        print("\nDEMO COMPLETA - TODOS LOS EJEMPLOS")
        print("="*50)

        parametros_demo = {
            'tam_poblacion': 30,
            'generaciones': 50,
            'prob_cruce': 0.7,
            'prob_mutacion': 0.2
        }

        print("Ejecutando todos los ejemplos con parámetros optimizados...")
        print("Esta demo ejecutará todas las opciones secuencialmente")

        confirmar = input("¿Continuar? (s/N): ").strip().lower()
        if confirmar not in ['s', 'si', 'y', 'yes']:
            print("Demo cancelada")
            return

        # Ejecutar cada opción
        opciones_a_ejecutar = [
            ("Maximización de función", lambda: ejecutar_opcion_1(parametros_demo)),
            ("Problema del viajante", lambda: ejecutar_opcion_2(parametros_demo)),
            ("Comparación de selecciones", ejecutar_comparacion_selecciones),
            ("Análisis de parámetros", ejecutar_analisis_parametros),
            ("Experimento personalizado", ejecutar_experimento_personalizado),
            # ----- NUEVO: ejecutar DEAP y mutación/cruce en modo auto -----
            ("Demostración DEAP", lambda: ejecutar_deap_ejemplo(auto=True)),
            ("Mutación y cruce", lambda: demostrar_mutacion_cruce(auto=True))
        ]

        for i, (nombre, funcion) in enumerate(opciones_a_ejecutar, 1):
            print(f"\n{'='*20} {i}/7: {nombre} {'='*20}")
            try:
                funcion()
                print(f" {nombre} completado")
            except Exception as e:
                print(f" Error en {nombre}: {str(e)}")

            if i < len(opciones_a_ejecutar):
                pausa_para_continuar("Presiona Enter para continuar con el siguiente ejemplo...")

        print("\n¡Demo completa finalizada!")
        print("Has visto todos los aspectos de los algoritmos genéticos")

    except Exception as e:
        print(f"Error en demo completa: {str(e)}")

def main():
    """Función principal del programa"""
    menu = MenuInteractivo()

    print("¡Bienvenido al Sistema Interactivo de Algoritmos Genéticos!")
    print("Este programa te permite explorar diferentes aspectos de los AGs")

    while True:
        try:
            menu.mostrar_menu()
            seleccion = menu.obtener_seleccion()

            if seleccion == '0':
                print("\n¡Gracias por usar el sistema de Algoritmos Genéticos!")
                print("¡Sigue explorando la evolución artificial!")
                break

            # Obtener parámetros para opciones que los requieren
            if seleccion in ['1', '2']:
                print(f"\nConfigurando parámetros para: {menu.opciones[seleccion]}")
                parametros = menu.obtener_parametros_basicos()
            else:
                parametros = None

            print(f"\nEjecutando: {menu.opciones[seleccion]}")

            # Ejecutar la opción seleccionada
            if seleccion == '1':
                ejecutar_opcion_1(parametros)
            elif seleccion == '2':
                ejecutar_opcion_2(parametros)
            elif seleccion == '3':
                ejecutar_comparacion_selecciones()
            elif seleccion == '4':
                ejecutar_analisis_parametros()
            elif seleccion == '5':
                ejecutar_experimento_personalizado()
            elif seleccion == '6':
                ejecutar_deap_ejemplo(auto=False)  # <-- ahora pide tus parámetros
            elif seleccion == '7':
                demostrar_mutacion_cruce(auto=False)  # <-- ahora es dinámica
            elif seleccion == '8':
                ejecutar_demo_completa()

            print(f"\n '{menu.opciones[seleccion]}' completado exitosamente")

        except KeyboardInterrupt:
            print("\n Operación interrumpida por el usuario")
            print("Regresando al menú principal...")
        except Exception as e:
            print(f"\n Error inesperado: {str(e)}")
            print("Intenta con diferentes parámetros o reporta el error")

        # Pausa antes de regresar al menú
        pausa_para_continuar("Presiona Enter para regresar al menú principal...")

def configurar_entorno():
    """Configurar el entorno de trabajo"""
    # Configurar numpy para reproducibilidad
    np.random.seed(None)  # Permitir aleatoriedad real

    # Configurar matplotlib
    plt.rcParams['figure.facecolor'] = 'white'
    plt.rcParams['axes.facecolor'] = 'white'
    plt.rcParams['font.size'] = 10

    # Configurar warnings
    warnings.filterwarnings('ignore', category=RuntimeWarning)

def mostrar_informacion_inicial():
    """Mostrar información sobre el programa"""
    print("\n" + "="*70)
    print("SISTEMA INTERACTIVO DE ALGORITMOS GENÉTICOS")
    print("="*70)
    print("CARACTERÍSTICAS DEL SISTEMA:")
    print(" • Algoritmos genéticos optimizados y modulares")
    print(" • Múltiples funciones objetivo para experimentar")
    print(" • Problema del viajante con visualización")
    print(" • Comparación de métodos de selección")
    print(" • Análisis de parámetros y su impacto")
    print(" • Integración con DEAP (librería profesional) con parámetros personalizables")
    print(" • Visualizaciones interactivas y educativas")
    print("\nBENEFICIOS:")
    print(" • Comprende los conceptos fundamentales de los AGs")
    print(" • Experimenta con diferentes configuraciones")
    print(" • Visualiza el proceso evolutivo paso a paso")
    print(" • Compara diferentes enfoques y estrategias")
    print("\n CONSEJOS:")
    print(" • Usa Ctrl+C para interrumpir cualquier operación")
    print(" • Cierra las ventanas de gráficas para continuar")
    print(" • Experimenta con diferentes parámetros")
    print("="*70)

# Ejecutar el programa
if __name__ == "__main__":
    try:
        configurar_entorno()
        mostrar_informacion_inicial()
        main()
    except KeyboardInterrupt:
        print("\n\n¡Programa terminado por el usuario!")
    except Exception as e:
        print(f"\nError crítico del programa: {str(e)}")
        print("Por favor reporta este error")
    finally:
        print("\n¡Gracias por usar el sistema de Algoritmos Genéticos!")
```
